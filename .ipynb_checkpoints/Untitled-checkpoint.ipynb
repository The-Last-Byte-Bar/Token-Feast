```python
from ergo_python_appkit.appkit import ErgoAppKit
from org.ergoplatform.appkit import Address, ErgoValue, Constants

# Initialize ErgoAppKit
node_url = "http://37.27.198.175:9053/"
explorer_url = "https://api.ergoplatform.com/"
appKit = ErgoAppKit(node_url, "mainnet", explorer_url, 'hashcream')

# Define addresses
minter_address = "9ebLdLKX7k6gAyruUPeqrT1jqKtscvFiW11EFmX5tcAvmwfZSsH"

def mint_tokens(token_name, token_description, token_amount, decimals=0):
    
    # Use the first input box's ID as the token ID
    token_id = unspent_boxes[0].getId().toString()
    
    # Create the minting output box
    mint_output = appKit.mintToken(
        value=int(1e7),  # 0.01 ERG
        tokenId=token_id,
        tokenName=token_name,
        tokenDesc=token_description,
        mintAmount=token_amount,
        decimals=decimals,
        contract=appKit.contractFromAddress(minter_address)
    )
    
    # Build and sign the transaction
    unsigned_tx = appKit.buildUnsignedTransaction(
        inputs=unspent_boxes,
        outputs=[mint_output],
        fee=int(1e6),
        sendChangeTo=Address.create(minter_address).getErgoAddress()
    )
    signed_tx = appKit.signTransactionWithNode(unsigned_tx)
    
    # Send the transaction
    tx_id = appKit.sendTransaction(signed_tx)
    print(f"Minting transaction sent. Transaction ID: {tx_id}")
    return token_id

def create_time_locked_box(token_id, token_amount, unlock_height):
    # Get input boxes containing the minted token
    inputs = appKit.boxesToSpend(minter_address, int(1e7), {token_id: token_amount})
    
    # Create the time-locked contract
    time_lock_script = f"HEIGHT > {unlock_height}"
    time_lock_contract = appKit.compileErgoScript(time_lock_script)
    
    # Create the time-locked output box
    time_locked_output = appKit.buildOutBox(
        value=int(1e7),  # 0.01 ERG
        tokens={token_id: token_amount},
        registers=None,
        contract=appKit.contractFromTree(time_lock_contract)
    )
    
    # Build and sign the transaction
    unsigned_tx = appKit.buildUnsignedTransaction(
        inputs=inputs,
        outputs=[time_locked_output],
        fee=int(1e6),
        sendChangeTo=Address.create(minter_address).getErgoAddress()
    )
    signed_tx = appKit.signTransactionWithNode(unsigned_tx)
    
    # Send the transaction
    tx_id = appKit.sendTransaction(signed_tx)
    print(f"Time-locked box created. Transaction ID: {tx_id}")

def spend_time_locked_box(token_id, token_amount):
    # Find the time-locked box
    time_locked_box = appKit.getUnspentBoxesFor(minter_address, withTokenId=token_id)[0]
    
    # Create the output box (sending tokens back to minter)
    output = appKit.buildOutBox(
        value=int(1e7),  # 0.01 ERG
        tokens={token_id: token_amount},
        registers=None,
        contract=appKit.contractFromAddress(minter_address)
    )
    
    # Build and sign the transaction
    unsigned_tx = appKit.buildUnsignedTransaction(
        inputs=[time_locked_box],
        outputs=[output],
        fee=int(1e6),
        sendChangeTo=Address.create(minter_address).getErgoAddress()
    )
    signed_tx = appKit.signTransactionWithNode(unsigned_tx)
    
    # Send the transaction
    tx_id = appKit.sendTransaction(signed_tx)
    print(f"Time-locked box spent. Transaction ID: {tx_id}")

# Usage example
if __name__ == "__main__":
    # Mint tokens
    token_id = mint_tokens("MyToken", "This is my custom token", 1000)
    
    # Create time-locked box (unlock after 2 blocks)
    current_height = appKit._ergoClient.execute(lambda ctx: ctx.getHeight())
    create_time_locked_box(token_id, 1000, current_height + 2)
    
    # Wait for 2 blocks...
    
    # Spend time-locked box
    #spend_time_locked_box(token_id, 1000)
```
